```python
from keras.preprocessing import image
import keras.utils as image # 쥬피트에서 사용 (코랩에서는 다른 것을 사용)
import matplotlib.pyplot as plt
%matplotlib inline
from keras.applications.vgg19 import VGG19, preprocess_input, decode_predictions

vggnet = VGG19(include_top=True, weights='imagenet',
               input_tensor=None, input_shape=None,
               pooling=None, classes=1000)
vggnet.summary()
```

```python
img = image.load_img('c:/users/isola/documents/python_study/dog.jpg', target_size=(224, 224))
plt.imshow(img)

x = image.img_to_array(img)
x = x.reshape((1, x.shape[0], x.shape[1], x.shape[2]))
x = preprocess_input(x)

preds = vggnet.predict(x)
print(decode_predictions(preds))

from PIL import Image

plt.imshow(img)
plt.show()
```
